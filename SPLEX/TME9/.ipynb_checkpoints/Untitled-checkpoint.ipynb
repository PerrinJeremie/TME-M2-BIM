___0. Importing libraries___

```python
import pandas as pd
import numpy as np
import pyAgrum as gum
import pyAgrum.lib.notebook as gnb
import pyAgrum.lib.bn2graph as bnPlot
import pyAgrum.lib.dynamicBN as gdyn
from IPython.display import Image
```

___1. Creating a network using PyAgrum___

First adding the nodes.

```python
v = {'G':gum.LabelizedVariable('Glycemia'),'I':gum.LabelizedVariable('Insulin'),\
             'O':gum.LabelizedVariable('Other_drugs'),'R':gum.LabelizedVariable('Remission')}
vID = {'G':0,'I':0,'O':0,'R':0}
BN = gum.BayesNet()
for key,val in v.items():
    vID[key] = BN.add(val)
print(BN)
```

Now adding the arcs and the conditionnal probabilities.

```python
BN.addArc(vID['G'],vID['I'])
BN.addArc(vID['G'],vID['O'])
BN.addArc(vID['I'],vID['R'])
BN.addArc(vID['O'],vID['R'])
```

```python
p=gum.Potential().add(BN.variable(vID['G'])).fillWith([1,1])
q1=gum.Potential().add(BN.variable(vID['I'])).add(BN.variable(vID['G'])).fillWith([9,1,1,9])
q2=gum.Potential().add(BN.variable(vID['O'])).add(BN.variable(vID['G'])).fillWith([7,3,25,75])
r=gum.Potential().add(BN.variable(vID['R'])).add(BN.variable(vID['I'])).add(BN.variable(vID['O'])).fillWith([1,9,7,3,4,6,9,1])
```

```python
BN.changePotential(vID['G'],p)
BN.changePotential(vID['I'],q1)
BN.changePotential(vID['O'],q2)
BN.changePotential(vID['R'],r)
```

```python
g = bnPlot.BN2dot(BN)
Image(g.create_png())
```

```python
gnb.showInference(BN,evs={})
```

Computing probabilities of remission when no drugs are taken.

```python
l = gum.LazyPropagation(BN)
l.addEvidence(vID['O'],0)
l.addEvidence(vID['I'],0)
l
```

```python
l.posterior(vID['R'])
```

Computing probabilities of remission if insulin prescribed and high glycemia

```python
l = gum.LazyPropagation(BN)
l.addEvidence(vID['G'],1)
l.addEvidence(vID['I'],1)
l.posterior(vID['R'])
```
