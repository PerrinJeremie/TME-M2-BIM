## TME 6 : Support Vector MachinesHyperparameter tuning and K-fold cross validation

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.datasets import make_blobs
from sklearn.datasets import make_moons
from sklearn import svm
from sklearn.model_selection import train_test_split
from sklearn.model_selection import KFold
import warnings; warnings.simplefilter('ignore')
```

```python

# First simulated data set
plt.figure(figsize=(14,4), dpi=80)
plt.subplot(131)
plt.title("Two informative features, one cluster per class", fontsize='small')
X1, Y1 = make_classification(n_samples=10000, n_features=2, n_redundant=0, n_informative=2,n_clusters_per_class=1)
plt.scatter(X1[:, 0], X1[:, 1], marker='o', c=Y1,s=20, edgecolor='k')


# Second simulated data set
plt.subplot(132)
plt.title("Three blobs", fontsize='small')
X2, Y2 = make_blobs(n_samples=10000, n_features=2, centers=3)
plt.scatter(X2[:, 0], X2[:, 1], marker='o', c=Y2, s=20, edgecolor='k')


# Third simulated data set
plt.subplot(133)
plt.title("Non-linearly separated data sets", fontsize='small')
X3, Y3 = make_moons(n_samples=10000, shuffle=True, noise=None, random_state=None)
plt.scatter(X3[:, 0], X3[:, 1], marker='o', c=Y3, s=20, edgecolor='k')
plt.show()

```

```python
valid_rate = 0.1
TV_array = []
for X,Y in [(X1,Y1),(X2,Y2),(X3,Y3)]:
    #XT,XV,YT,YV
    TV_array.append(train_test_split(X, Y, test_size=valid_rate))
    
```

```python

# First simulated data set
plt.figure(figsize=(14,4), dpi=80)
plt.subplot(131)
plt.title("Two informative features, one cluster per class", fontsize='small')
plt.scatter(TV_array[0][1][:, 0], TV_array[0][1][:, 1], marker='o', c=TV_array[0][3],s=20, edgecolor='k')


# Second simulated data set
plt.subplot(132)
plt.title("Three blobs", fontsize='small')
plt.scatter(TV_array[1][1][:, 0], TV_array[1][1][:, 1], marker='o', c=TV_array[1][3], s=20, edgecolor='k')


# Third simulated data set
plt.subplot(133)
plt.title("Non-linearly separated data sets", fontsize='small')
plt.scatter(TV_array[2][1][:, 0], TV_array[2][1][:, 1], marker='o', c=TV_array[2][3], s=20, edgecolor='k')
plt.show()
```

```python
def doKFold(X,Y,k,c,ker):
    kf = KFold(n_splits=k)
    acc = 0
    model = svm.SVC(C = c,kernel = ker)
    for train_index, test_index in kf.split(X):
        X_train, X_test = X[train_index], X[test_index]
        Y_train, Y_test = Y[train_index], Y[test_index]
        model.fit(X_train, Y_train)
        acc += model.score(X_test,Y_test)
    return acc/k

C = np.arange(0.1,10,0.5)
hyp_par = ["linear","poly","rbf"]
chosen_hyp_par = []

for val in TV_array:
    bc = 0
    bhyp = ""
    bacc = -1
    for c in C:
        for ker in hyp_par:
            acc = doKFold(val[1],val[3],10,c,ker)
            if acc > bacc:
                bacc = acc
                bhyp = ker
                bc = c
    chosen_hyp_par.append([bc,bhyp])
```

```python
chosen_hyp_par
```

```python
def doKFold_all(X,Y,k,c,ker):
    kf = KFold(n_splits=k)
    acc = []
    model = svm.SVC(C = c,kernel = ker)
    for train_index, test_index in kf.split(X):
        X_train, X_test = X[train_index], X[test_index]
        Y_train, Y_test = Y[train_index], Y[test_index]
        model.fit(X_train, Y_train)
        acc.append(model.score(X_test,Y_test))
    return acc

accuracies = []
for i in range(3):
    val = TV_array[i]
    c = chosen_hyp_par[i][0]
    ker = chosen_hyp_par[i][1]
    accuracies.append(doKFold_all(val[0],val[2],10,c,ker))
    
```

```python
accuracies
```

```python
plt.figure(figsize=(14,4), dpi=80)
plt.subplot(131)
plt.title('For First training set')
plt.boxplot(accuracies[0])

plt.subplot(132)
plt.title('For Second training set')
plt.boxplot(accuracies[1])

plt.subplot(133)
plt.title('For Third training set')
plt.boxplot(accuracies[2])
plt.show()
```

```python
def maxindex(t):
    if type(t) != np.ndarray:
        return (t,t > 0)
    else:
        i = 0
        maxi = t[0]
        for j in range(1,3):
            if t[j] > maxi:
                i = j
                maxi = t[j]
        return (maxi,i)


for k in range(3):
    clf = svm.SVC(kernel=chosen_hyp_par[k][1], C=chosen_hyp_par[k][0])
    clf.fit(TV_array[k][0], TV_array[k][2])

    plt.figure()
    plt.clf()
    plt.scatter(TV_array[k][0][:, 0], TV_array[k][0][:, 1], c=TV_array[k][2], zorder=10, cmap=plt.cm.Paired,
                edgecolor='k', s=20)

    # Circle out the test data
    plt.scatter(TV_array[k][1][:, 0], TV_array[k][1][:, 1], s=80, facecolors='none',
                zorder=10, edgecolor='k')

    plt.axis('tight')
    x_min = TV_array[k][0][:, 0].min()
    x_max = TV_array[k][0][:, 0].max()
    y_min = TV_array[k][0][:, 1].min()
    y_max = TV_array[k][0][:, 1].max()

    XX, YY = np.mgrid[x_min:x_max:200j, y_min:y_max:200j]
    Z = clf.decision_function(np.c_[XX.ravel(), YY.ravel()])
    
    # Put the result into a color plot
    Zm = [maxindex(z) for z in Z ]
    Z = np.array([z[0] for z in Zm])
    Zm = np.array([z[1] for z in Zm])
    Z = Z.reshape(XX.shape)
    Zm = Zm.reshape(XX.shape)
    plt.pcolormesh(XX, YY, Zm, cmap=plt.cm.Paired)
    plt.contour(XX, YY, Z, colors=['k', 'k', 'k'],
                linestyles=['--', '-', '--'], levels=[-.5, 0, .5])

    plt.title(chosen_hyp_par[k][1])
plt.show()
```

The support vectors are the data points which define the decision boundary.
That is why a SVM is very light because all points which are not support vectors can be forgotten for the decision process.
They are the points which mximize the margin.
